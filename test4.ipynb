```python
#多线程打开多个好友窗口进行消息监听
from concurrent.futures import ThreadPoolExecutor
from pyweixin import Navigator,Monitor
#先打开所有好友的独立窗口
dialog_windows=[]
friends=['Hello,Mr Crab','Pywechat测试群','一家人']
durations=['1min']*len(friends)
#不添加其他参数,比如save_photos,这些操作涉及键鼠,无法多线程
for friend in friends:
    dialog_window=Navigator.open_seperate_dialog_window(friend=friend,window_minimize=True,close_weixin=True)
    dialog_windows.append(dialog_window)
with ThreadPoolExecutor(max_workers=len(friends)) as pool:
    results=pool.map(lambda args: Monitor.listen_on_chat(*args),list(zip(dialog_windows,durations)))
for friend,result in zip(friends,results):
    print(friend,result)
```

```python
import os
def find_files_scandir(directory):
    """递归查找文件"""
    files=[]
    try:
        with os.scandir(directory) as entries:
            for entry in entries:
                if entry.is_file():
                    files.append(entry.path)
                elif entry.is_dir():
                    # 递归搜索子目录
                    files.extend(find_files_scandir(entry.path))
    except PermissionError:
        print(f"错误o(╥﹏╥)o：无权限访问目录: {directory}")
    files=list(set(files))
    return files
find_files_scandir(os.getcwd())
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from tqdm import tqdm

# 加载数据集
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # 初始化权重参数
        self.w1 = nn.Parameter(torch.from_numpy(np.random.randn(6, 4)).float())
        self.w2 = nn.Parameter(torch.from_numpy(np.random.randn(8, 6)).float())
        self.w3 = nn.Parameter(torch.from_numpy(np.random.randn(3, 8)).float())
        
        # 初始化偏置参数
        self.b1 = nn.Parameter(torch.zeros(6).float())
        self.b2 = nn.Parameter(torch.zeros(8).float())
        self.b3 = nn.Parameter(torch.zeros(3).float())
    
    def forward(self, x):
        # 第一层：4个特征 -> 6个神经元
        f1 = torch.sigmoid(torch.matmul(x, self.w1.T) + self.b1)
        # 第二层：6个神经元 -> 8个神经元
        f2 = torch.sigmoid(torch.matmul(f1, self.w2.T) + self.b2)
        # 输出层：8个神经元 -> 3个输出（注意：没有激活函数，使用线性输出）
        o = torch.matmul(f2, self.w3.T) + self.b3  # 移除sigmoid，CrossEntropyLoss会处理
        return o

if __name__ == '__main__':
    # 初始化网络
    _net = Net()
    
    # 使用交叉熵损失（适用于分类问题）
    _loss_fn = nn.CrossEntropyLoss()
    
    # 优化器，降低学习率
    _opt = optim.SGD(_net.parameters(), lr=0.01, momentum=0.9)
    
    # 准备数据
    _x = torch.from_numpy(iris.data).float()  # 形状: (150, 4)
    
    _y = torch.from_numpy(iris.target).long() # 形状: (150,)
    # 注意：CrossEntropyLoss期望的是类别索引（long类型），不是one-hot编码
    
    # 训练循环
    for i in tqdm(range(10000)):
        _opt.zero_grad()
        _py = _net(_x)           # 前向传播，输出形状: (150, 3)
        _loss = _loss_fn(_py, _y)  # 计算损失
        _loss.backward()           # 反向传播
        _opt.step()                # 更新参数
    
    # 打印训练结果（只在训练结束后打印一次）
    print("\n训练完成!")
    print(f"最终损失为: {_loss.item():.6f}")
    
    # 可以添加准确率计算
    with torch.no_grad():
        predictions=torch.argmax(_net(_x),dim=1)
        accuracy=(predictions==_y).float().mean()
        print(f"训练集准确率:{accuracy.item()*100:.2f}%")
    
    # 查看参数（可选）
    print("\n第一层权重:")
    print(_net.w1.data)
    print("\n第一层偏置:")
    print(_net.b1.data)
```
